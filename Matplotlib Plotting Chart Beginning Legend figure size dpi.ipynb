```python
import matplotlib.pyplot as plt
#There two ways of using matplotlib, functional method and object oriented method
#The object oriented method is the better way of plotting
```

```python
%matplotlib inline 
#This is just to show matplotlib, the plotting in our jupyter notebook
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Using the functional method
plt.plot(x,y)
#plt.show() is for non jupyter notebook to show the chart/graph
```

```python
plt.xlabel('Normz')
plt.ylabel('Squares')
plt.title('Title S')
#has to be plotted together with the plot
```

```python
plt.plot(x,y)
plt.xlabel('Normz')
plt.ylabel('Squares')
plt.title('Title S')
```

```python
#Using subplot there are 3 argument or more i guess
#The number of row, the number of columns, the plot number
plt.subplot(1,2,1)
plt.plot(x,y,'r')
#This says one row by two columns and plot number 1, plot x by y with color red for r

plt.subplot(1,2,2)
plt.plot(y,x,'b')
#Same as above, note that row is 1 n column is 2
#Which shows 1 row of charts with 2 columns of chart if it was 2 rows n 1 column then they would be stacked on top each other n not side by side
```

```python
#Object Oriented Method
fig = plt.figure()
#create a figure object
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#on the fig obj add a set of axes
#Add axes takes in 4 arguments, left, bottom, width n height arg that range from 0 to 1
axes.plot(x,y)
#Then plot x n y on that axes
axes.set_xlabel('Normz')
axes.set_ylabel('Squares')
axes.set_title('Title S')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.45,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Large Plot')

axes2.plot(y,x)
axes2.set_title('Small Plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.4,0.8])
axes2 = fig.add_axes([0.5,0.1,0.4,0.8])

axes1.plot(x,y,'r')
axes1.set_title('Large Plot')

axes2.plot(y,x,'y')
axes2.set_title('Small Plot')
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
#Using subplots with OO(Object Oriented)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1)
#you can specify the number of rows n columns above
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#you can specify the number of rows n columns above
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
#.tight_layout is recommended to put at the end of your code
#This fixes the issue of overlapping
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#axes is an array object so can be iterated through
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
#you can also index the axes object
axes[0].plot(x,y)
#plot on index zero of axes
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('1st Plot')
axes[1].plot(y,x)
axes[1].set_title('2nd Plot')
#All you are doing is grabbing the plot you want to edit using indexing
```

```python
#figure size n DPI
#figure size is just width n height in inches
#DPI is dot per inch or pixel per inch, so bigger number means larger graph
#fig = plt.figure(figsize=(3,2), dpi=100)
fig = plt.figure(figsize=(6,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(6,3))

axes[0].plot(x,y)
axes[0].set_title('1st Plot')
axes[1].plot(y,x)
axes[1].set_title('2nd Plot')
plt.tight_layout()
```

```python
#Saving a figure into the format you want, picture, pdf etc
fig.savefig('Dimz_Chart.jpg',dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.plot(y,x)
```

```python
#Legends gives the labelling of multiple lines in a chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X2')
ax.plot(x,x**3, label='X3')

ax.legend()
#.legend(loc=0), location is where the legend show appear in the chart
#0 is best, 1 is upper right, 2 is upper left, 3 low l, 4 low r, center is 10, check doc for others
#.legend(loc=(0.1,0.1)) can use a tuple like in axes, left bottom...
#label = in plot arg is needed for legends to work
```

```python
fig.savefig('Dimz_Ax_Chart.jpg',dpi=150)
```

```python
#Color
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple')
#Color code can be string(purple, orange etc) hex code(#FF8C00) or single string(r, b etc)
```

```python
#line width
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', linewidth=12, alpha=0.5)
#alpha is the visibility or see through
#linewidth can also just be lw
```

```python
#line style
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', lw=5, linestyle='dashed')
#line style(ls) can be . or ., or : or dashdotted or a bunch check docs to see them
```

```python
#Marker
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple', lw=1, alpha=0.8, ls='-', marker='o', markersize=20, markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue')
#markers marks the point in the line that the x versus y meet
#There are alot of arguments you can add with marker
#markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='blue'
```

```python
#Axis limit
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue', lw=2, alpha=0.9, ls='--')
ax.set_xlim([0,1])#upper bound n lower bound of x axis
ax.set_ylim([0,2])#upper bound n lower bound of y axis
```
