```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25)
#can add cmap='coolwarm' too
#hexbin is like scatter plot but in hexagons
#gridsize for increasing the hexagon sizes, they get darker with the frequency of times appeared
```
